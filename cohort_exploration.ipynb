```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
metadata = pd.read_excel('/Users/clairez/Desktop/BENG203_Final_Project/Data/toden_metadata.xlsx')
metadata.rename(columns={'Run': 'Sample_id'}, inplace=True)
metadata
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(metadata["Age"].dropna(), kde=True, bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=metadata, x="Disease", order=metadata["Disease"].value_counts().index)
plt.title("Disease Frequency")
plt.xlabel("Disease")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
